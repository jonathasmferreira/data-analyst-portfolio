## 🧠 Conceitos-chave

- **`pd.read_csv()`** aceita caminho local **ou URL** direto. Útil para datasets públicos.
- **`.head(n)` / `.tail(n)`** mostram as primeiras / últimas `n` linhas (padrão: 5).
- **`.shape`** é um atributo (sem parênteses), retorna tupla `(linhas, colunas)`.
- **`.info()`** mostra estrutura (tipos, nulos); **`.describe()`** mostra estatísticas numéricas.
- **Seleção de colunas**: `df['col']` retorna **Series** (1 coluna); `df[['col1', 'col2']]` retorna **DataFrame** (mesmo se for 1 coluna só, mas dentro de lista).

## ⚠️ Erros comuns

1. **Usar lista simples em vez de dupla** — `df['Name', 'Age']` (sem o segundo par de colchetes) dá erro. O jeito certo é `df[['Name', 'Age']]`.
2. **Arquivo não encontrado** — se o CSV não está na pasta certa, dá erro. Confirme com `import os; print(os.getcwd())` em qual pasta o notebook está rodando.
3. **Caracteres estranhos no texto** — alguns datasets brasileiros precisam de `encoding='latin1'` ou `encoding='utf-8-sig'` na hora de ler.
4. **Esquecer que `.shape` não tem parênteses** — é uma propriedade, não uma função. `df.shape()` dá erro.
5. **Achar que `.describe()` mostra tudo** — por padrão só mostra colunas numéricas. Para incluir texto, use `df.describe(include='all')`.
